```python
def isprime(num):
  count=0
  for i in range(2,num):
    if num%i==0:
      count=count+1
  if count==0:
    print("number is prime here:",num)
  else:
    print("number is not prime:",num)

num=int(input("enter the number:"))
isprime(num)
```

```python
def odd_even(num):
  if num%2==0:
    print("even",num)
  else:
    print("odd",num)

num=int(input("enter the number"))
odd_even(num)
```
